# Here we are going to calculate the average sprint speed

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks

df = pd.read_csv('matrix_player_6.csv')
df = df.fillna(0)
```

```python
length = int(len(df['timeLine']) * 0.02)
temp = df.iloc[:length, :]
```

```python
for index in range(length) :
    if df.loc[index, [' frameRotationalSpeedZ ']].values > 5 or df.loc[index, [' frameRotationalSpeedZ ']].values < -5 :
        temp = temp.drop(index, axis=0)
```

```python
peakDistance = 1000
```

```python
peaks, _ = find_peaks(temp.loc[:len(temp), 'wheelRotationalSpeedX '], distance=peakDistance)

meanPeak = peaks.mean()
print(meanPeak)
```
