```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

df_video = pd.read_csv('Ned - DUI Game 1.csv')
print(df_video)
print('---------------------------------------------------------------')
df_player = pd.read_csv('matrix_player_6.csv')
print(df_player)

!chmod 660 predicting_sprint.ipynb
```

Below i will run a few plots.

```python
df_player = df_player.fillna(0)
print(df_player)
```

```python

sprints = [0]*770883
for x in range(770883) :
    spd = df_player.loc[x,' frSpeed ']
    acc = df_player.loc[x,'frAcc ']
    if spd > 1 and acc > 10 :
        sprints[x] = 1

df_player["IsSprinting"] = sprints
#print(df_player.loc[df_player["IsSprinting"] == 1])
X= df_player[['frAcc ',' frSpeed ']]

#X= df_player['frAcc ']
    
targets= df_player['IsSprinting'].values
print(X)
print(X.shape)
plt.clf()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, targets, test_size=0.5, random_state=42)
```

```python
print(sum(y_train))
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
pred = gnb.predict(X_test)
print(pred)
```

```python
y_test
```

```python
gnb.score(pred.reshape(-1,1),y_test)
```
